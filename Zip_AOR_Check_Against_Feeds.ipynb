```python
# @title Run Accuray Check
import pandas as pd
import re

def extract_zip(address: str) -> str:
    """
    Extracts the last matching zip code from the given address.

    Args:
        address (str): The address to extract the zip code from.

    Returns:
        str: The extracted zip code, or raises a ValueError if not found.
    """
def extract_zip(address):
    zip_pattern = r'\b\d{5}\b'
    matches = re.findall(zip_pattern, address)
    return matches[-1] if matches else 'No match found'

def filter_market(select_dma: str, aor: pd.DataFrame) -> pd.DataFrame:
    """
    Filters the market data based on the given DMA and LMA indicator.

    Args:
        select_dma (str): The DMA to filter by.
        aor (pd.DataFrame): The market data to filter.

    Returns:
        pd.DataFrame: The filtered market data.
    """
    filtered_market = aor[(aor['DMA Name'] == select_dma) & (aor['LMA Indicator'] == 'Enrolled')]
    print(f"Market to check: {select_dma}")
    print(f"There are a total of {len(filtered_market['Dealer Name'].unique())} enrolled dealers.")
    print("----------------------------------")
    filtered_market.loc[:,'Zip Percent of AGSSA'] = filtered_market['Zip Percent of AGSSA'].replace('< 1%', 0.01).astype(float)
    filtered_market.loc[:, 'Zip Code'] = filtered_market['Zip Code'].astype(str)
    for dealer in filtered_market['Dealer Name'].unique():
        print(dealer)
    return filtered_market

def dealer_ided(filtered_market: pd.DataFrame, dealer_location_zip: str) -> str:
    """
    Finds the BAC for the given dealer location zip.

    Args:
        filtered_market (pd.DataFrame): The filtered market data.
        dealer_location_zip (str): The dealer location zip to find the BAC for.

    Returns:
        str: The BAC for the given dealer location zip, or an empty string if not found.
    """
    if dealer_location_zip not in filtered_market['Zip Code'].values:
        print("Dealer not found")
        return ''
    tmp_dealer_id = filtered_market[filtered_market['Zip Code'] == dealer_location_zip]
    max_index = filtered_market.loc[filtered_market['Zip Code'] == dealer_location_zip, 'Zip Percent of AGSSA'].idxmax()
    print(filtered_market.loc[max_index])
    print("----------------------------------")
    print(f"Found BAC for {filtered_market.loc[max_index]['Dealer Name']} BAC: {filtered_market.loc[max_index]['BAC']}")
    return filtered_market.loc[max_index]['BAC']

def dealer_aor_zips(filtered_market: pd.DataFrame, BAC: str) -> list:
    """
    Finds the AOR zips for the given BAC.

    Args:
        filtered_market (pd.DataFrame): The filtered market data.
        BAC (str): The BAC to find the AOR zips for.

    Returns:
        list: The AOR zips for the given BAC.
    """
    return list(filtered_market[filtered_market['BAC'] == BAC]['Zip Code'].unique())

def convert_geo_zip_to_list(geo_zip: str) -> list:
    """
    Converts the geo zip to a list.

    Args:
        geo_zip (str): The geo zip to convert.

    Returns:
        list: The converted geo zip.
    """
    if geo_zip == 'No match found':
        return []
    elif isinstance(geo_zip, list):
        return geo_zip
    elif isinstance(geo_zip, str):
        return geo_zip.split(',')

def lma_whole_accuracy_check(lma: pd.DataFrame):
    """
    Performs a whole accuracy check on the LMA data.

    Args:
        lma (pd.DataFrame): The LMA data to check.
    """
    tmp_lma = lma[lma['zip'] != 'No match found']
    if set(tmp_lma['Match'].unique()) == {True}:
        print("✅ Accuracy check passed")
        print('Existing feed does not require updates')
    elif set(tmp_lma['Match'].unique()) == {False}:
        print("‼️ Accuracy check failed")
        print('Existing feed requires updates:')
        print(tmp_lma[tmp_lma['Match'] == False])

filtered_market = filter_market(select_dma, aor)

# Address NAN dealer addresses
lma['DealerAddress1'] = lma['DealerAddress1'].fillna('')
lma['zip'] = lma['DealerAddress1'].apply(extract_zip)
tmp_dealer_indx = lma[['DealerName', 'zip']]
tmp_dealer_indx = tmp_dealer_indx[tmp_dealer_indx['zip'] != 'No match found']
tmp_dealer_indx.loc[:, 'zip'] = tmp_dealer_indx['zip'].astype(str)

# Find BAC for each dealer
lma['BAC'] = lma['zip'].apply(lambda x: dealer_ided(filtered_market, x))

# Find dealers without BAC
dealers_without_bac = list(lma[lma['BAC'] == ''][['DealerName', 'DealerAddress1']]['DealerName'].unique())
print("----------------------------------")
print("Dealers without BAC")
[print(item) for item in dealers_without_bac if not None]

# Find AOR zips for each dealer
lma['AOR zips'] = lma['BAC'].apply(lambda x: dealer_aor_zips(filtered_market, x))

# Convert geo zip to list
lma.loc[:,'Geo ZIP'] = lma['Geo ZIP'].apply(convert_geo_zip_to_list)

# Find matching zip codes
lma['Match'] = lma.apply(lambda row: set(row['AOR zips']) == set(row['Geo ZIP']), axis=1)

# run check against matched dealer zip codes
lma_whole_accuracy_check(lma)

# set temporary file name for output & download
tmp_file_name = f"{select_dma}_accuracy_check.csv"

# Save report to CSV
lma.to_csv(tmp_file_name, index=False)

# Download to local system
files.download(tmp_file_name)

#if __name__ == "__main__":
#     main()
```
